Dalam praktik data mining, missing values merujuk pada keadaan di mana nilai atau entri dalam dataset tidak tersedia atau tidak terisi. Ini adalah situasi yang sering terjadi dalam pengumpulan data dan bisa disebabkan oleh berbagai faktor, termasuk kesalahan dalam pengumpulan data, ketidakakuratan pengukuran, atau kekurangan informasi.

## menggunakan knn

### pengertian knn

KNN (K-Nearest Neighbor) adalah salah satu algoritma dalam machine learning yang digunakan untuk melakukan klasifikasi pada data. Algoritma ini bekerja dengan mencari kategori terdekat dari data yang baru diberikan berdasarkan jarak euclidean atau jarak manhattan. Kategori yang dipilih adalah kategori yang paling banyak muncul di antara tetangga terdekat dari data yang baru diberikan.

Rumus knn:

Rumus dasar untuk algoritma K-Nearest Neighbors (KNN) melibatkan beberapa tahapan sebagai berikut:

1. Perhitungan Jarak: Tahap ini mencakup menghitung jarak antara titik data baru dengan setiap titik data dalam dataset. Berbagai metrik jarak seperti Euclidean, Manhattan, atau metrik kustom lainnya bisa digunakan tergantung pada kebutuhan dan preferensi masalah.

2. Penentuan Tetangga Terdekat: Setelah jarak dihitung, langkah selanjutnya adalah memilih K titik data dengan jarak terdekat dari titik data yang baru. Nilai K biasanya sudah ditentukan sebelumnya sebagai parameter dan merupakan bilangan bulat positif.

3. Proses Voting atau Pemilihan Mayoritas: Dalam konteks klasifikasi, setelah K tetangga terdekat ditemukan, dilakukan proses voting untuk menentukan kelas mayoritas di antara mereka. Untuk regresi, nilai target bisa berupa rata-rata atau median dari nilai K tetangga terdekat.

4. Memberikan Prediksi: Langkah terakhir adalah memberikan prediksi untuk titik data baru berdasarkan hasil dari proses voting atau nilai rata-rata/median dari tetangga terdekat, sesuai dengan jenis masalah (klasifikasi atau regresi).

# Deteksi Outlier (lokal Outlier Factor)

## konseplokal outlier faktor

Konsep Local Outlier Factor (LOF) adalah metode dalam deteksi anomali yang digunakan untuk mengidentifikasi titik data yang merupakan outlier berdasarkan tingkat kepadatan lokalnya. LOF membandingkan kepadatan titik data dengan kepadatan tetangga-tetangganya untuk menentukan apakah titik data tersebut anomali atau tidak.

1. K-neighbors: K-neighbors dari sebuah titik data adalah kumpulan dari k titik data terdekat ke titik tersebut. Dalam hal ini, "k" adalah parameter yang ditentukan sebelumnya. Jadi, jika kita memiliki sebuah titik data P dan kita ingin menentukan 5-neighbors dari P, maka kita akan mencari lima titik data terdekat dari P.

<img src="image-20240321-043201.png" width="" align="" />

2. Reachability distance (RD)

 didefinisikan sebagai jarak K maksimum Xj dan jarak antara Xi dan Xj. Ukuran jarak bersifat khusus untuk masalah (Euclidean, Manhattan, dll.)

<img src="image-20240321-043222.png" width="" align="" />

3. Local reachability density (LRD) adalah kebalikan dari rata-rata jarak jangkauan A dari tetangga-tetangganya. Semakin besar jarak jangkauan rata-ratanya (yaitu, tetangga-tetangga yang jauh dari titik tersebut), semakin sedikit kepadatan titik yang ada di sekitar titik tertentu. Ini mengindikasikan seberapa jauh suatu titik dari kelompok titik terdekatnya. Nilai LRD yang rendah menunjukkan bahwa cluster terdekat berada jauh dari titik tersebut.

4.Local Outlier Factor (LOF) adalah ukuran yang digunakan untuk mengidentifikasi pencilan (outlier) dalam data. Ini dilakukan dengan membandingkan rata-rata Local Reachability Density (LRD) dari K tetangga suatu titik dengan LRD dari titik tersebut sendiri.

<img src="image-20240321-043238.png" width="" align="" />

Jika sebuah titik adalah bagian dari data yang umum (inlier), rata-rata LRD dari tetangganya akan hampir sama dengan LRD dari titik itu sendiri (karena kepadatan titik dan tetangganya hampir sama). Dalam hal ini, LOF akan mendekati 1. Namun, jika sebuah titik adalah pencilan, LRD dari titik tersebut akan lebih rendah daripada rata-rata LRD dari tetangganya. Hal ini akan menghasilkan nilai LOF yang tinggi.

<img src="image-20240321-043256.png" width="" align="" />

Umumnya, ketika nilai LOF > 1, titik tersebut dianggap sebagai pencilan. Namun, ini tidak selalu benar. Misalnya, jika kita mengetahui bahwa hanya ada satu pencilan dalam data, kita dapat mengambil nilai LOF maksimum dari semua titik dan mempertimbangkan titik yang sesuai dengan nilai LOF maksimum tersebut sebagai pencilan.

## menghitung menggunakan python (Scikit Learn)

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

X = [[4, 10, 4], [5, 11, 6], [6, 15, 3], [10, 11, 12], [14, 11, 13], [12, 15, 11]]
X = np.array(X)
clf = LocalOutlierFactor(n_neighbors=3, contamination='auto', p=2) 
y_pred = clf.fit_predict(X)
outlier_mask = y_pred == -1
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c='b', label='Data Points')
plt.scatter(X[outlier_mask, 0], X[outlier_mask, 1], c='r', label='Outliers')
plt.title('Local Outlier Factor (LOF)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
